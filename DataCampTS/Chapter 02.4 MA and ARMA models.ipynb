```python
# simulate a AR process
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')

from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
# simulate a MA process
from statsmodels.tsa.arima_process import ArmaProcess
ar = np.array([1])
ma = np.array([1, 0.5])
MA_object = ArmaProcess(ar, ma)
simulated_data = MA_object.generate_sample(nsample=1000)
plt.plot(simulated_data)
```

```python
# estimate an MA model
from statsmodels.tsa.arima_model import ARMA
mod = ARMA(simulated_data, order = (0, 1))
result = mod.fit()
```

```python
# forecasting an MA model
mod = ARMA(simulated_data, order=(0,1))
res = mod.fit()
res.plot_predict(start=990, end=1010)
plt.show()
```

```python
# AR 
ma = [0.8 for i in range(30)]
ar = np.array([1])
AR_object = ArmaProcess(ar, ma)
simulated_data = AR_object.generate_sample(nsample = 5000)
```

```python
plot_acf(simulated_data, lags=30)
plt.show()
```

```python
# MA
ma = [0.8**i for i in range(30)]
ar = np.array([1])
AR_object = ArmaProcess(ar, ma)
simulated_data = AR_object.generate_sample(nsample = 5000)
```

```python
plot_acf(simulated_data, lags=30)
plt.show()
```
