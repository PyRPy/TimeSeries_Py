```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
data = [446.6565,  454.4733,  455.663 ,  423.6322,  456.2713,  440.5881, 425.3325,  
        485.1494,  506.0482,  526.792 ,  514.2689,  494.211 ]
index= pd.date_range(start='1996', end='2008', freq='A')
oildata = pd.Series(data, index)
```

```python
# simple exponential smoothing
```

```python
ax=oildata.plot()
ax.set_xlabel("Year")
ax.set_ylabel("Oil (millions of tonnes)")
plt.show()
print("Figure 7.1: Oil production in Saudi Arabia from 1996 to 2007.")
```

```python
fit = SimpleExpSmoothing(oildata).fit(smoothing_level=0.2,optimized=False)
fcast = fit.forecast(3).rename(r'$\alpha=0.2$')
```

```python
ax = oildata.plot(marker='o', color='black', figsize=(7.5,5))
fcast.plot(marker='o', ax=ax, color='blue', legend=True)
fit.fittedvalues.plot(marker='o', ax=ax, color='blue')
plt.show()
```
